```python
import numpy as np
import pandas as pd
from dialogloader import *
from styler import *
```

```python
TRAIN_FILE = "train_final.json"
TEST_FILE = "newbies_final.csv"

#load train json data
data = load_json(TRAIN_FILE)
df = pd.DataFrame.from_dict(data).set_index("dialogId")

#load output csv format 
df_test = pd.read_csv("newbies_final.csv", header=None)
df_test.columns=["dialogId", "quality_alice_test", "quality_bob_test"]
df_test.set_index("dialogId", inplace=True)
#join DFs
df2 = pd.merge(df, df_test, how='left', left_index=True, right_index=True)
df2["quality_alice_diff"] = df2["quality_alice"] - df2["quality_alice_test"]
df2["quality_bob_diff"] = df2["quality_bob"] - df2["quality_bob_test"]

DialogStyler(df2.sort_values(["quality_alice_diff", "quality_bob_diff"]) \
             [["context", "quality_alice_diff", "quality_bob_diff", "text"]][:100],table_styles=magnify())
```
